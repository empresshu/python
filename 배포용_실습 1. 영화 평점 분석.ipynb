```python
movies2= movies.set_index('영화제목')
movies2.head(10)
```

```python
pd.concat([여성인기영화, movies2['장르']],axis=1, sort = True, join = 'inner')
```

```python
여성인기장르 = 여성인기영화['장르'].str.split('|',expand=True)
```

```python
인기장르 = 여성인기장르[0].value_counts().add(여성인기장르[1].value_counts(),fill_value=0)
for col in [2,3,4]:
    인기장르 = 인기장르.add(여성인기장르[col].value_counts(), fill_value = 0)
인기장르.sort_values(ascending = False)
```

```python
# 전체 평점의 개수 500개 이상 -> 남자평점평균, 여자 평점평균의 차이가 큰 10개의 영화 
data3 = data[data.성별 == 'F']
여자평점평균 = data3.pivot_table(index='영화제목', values = '평점', aggfunc = 'mean')  
data3_1 = data[data.성별 =='M']
남자평점평균 = data3_1.pivot_table(index='영화제목', values = '평점', aggfunc = 'mean')
차이 = abs(여자평점평균['평점'] - 남자평점평균['평점'])
```

```python
abs(여자평점평균['평점'] - 남자평점평균['평점'])
```

```python
평점개수 = data.pivot_table(index = '영화제목', values = '평점', aggfunc = ('count'))
```

```python
평점오백이상 = 평점개수[평점개수['평점']>=500]
```

```python
평점오백이상_평점차= pd.merge(평점오백이상,차이, how='left', left_on = '영화제목', right_on = '영화제목')
```

```python
남자와여자의호불호가크게갈리는영화 =  평점오백이상_평점차.sort_values(by = '평점_y', ascending = False).head(10)
남자와여자의호불호가크게갈리는영화
```

## 5. (실습) 연령대 별로 영화 평점 분석하기

```python
data.head(20)
```

```python
def 연령대생성 (x):
    if x <10 :
        return '10세 미만'
    elif x<20 :
        return '10대'
    elif x<30 :
        return '20대'
    elif x<40 :
        return '30대'
    elif x<50 :
        return '40대'
    elif x<60 :
        return '50대'
    else:
        return '60대 이상'
```

```python
data.연령.apply(연령대생성)
```

```python
data['연령대']=data.연령.apply(연령대생성)
```

```python
data.pivot_table(index='영화제목',columns='연령대', values = '평점', aggfunc=('count','mean'), fill_value = '-')
```
